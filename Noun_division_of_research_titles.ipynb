```python
# Janomeのインストール
#!pip install janome | tail -n 1
!pip install janome

# Janomeのロード
from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
from janome.charfilter import *
from janome.tokenfilter import *

import numpy as np
import pandas as pd
```

```python
test_data = pd.read_csv("test.csv") # タイトルデータ

#test_data2 = test_data.dropna(how='all')

# Tokenneizerインスタンスの生成 
t = Tokenizer()
```

```python
new_titles = []
```

```python
for i, title in test_data.iterrows():
  #print('title =',title['title'])
  tt = title['title']
  tokens = t.tokenize(tt)
  #print(tt)
  a = Analyzer(token_filters=[CompoundNounFilter()])
  new_title = ''
  for token in a.analyze(tt):
    if token.part_of_speech.split(',')[0] in['名詞']:
      #print(token.base_form)
      new_title += token.base_form + ' '
  new_titles.append(new_title)
'''
#print(new_title)
'''
```
